```python
# analysys:
# --tweets per tag--

query = """select hashtag.phrase, count(*) 
    from tweet left join hashtag on searched_hashtag_id::bigint = hashtag.id 
    group by hashtag.phrase"""

cur.execute(query)
rows = cur.fetchall()

print "Quantity:   \tTag:"
for row in rows:
    print "   ", row[1], "  \t", row[0]
```

```python
# analysys:
# -- retweets  per tweet--

query = """select sub.retweets, count(*) as num_of_tweets
from (select tweet_id, count(*) as retweets from
re_tweet group by tweet_id) as sub
group by sub.retweets
order by sub.retweets desc"""

cur.execute(query)
rows = cur.fetchall()

print "Tweets:   \tNumber of retweets:"
for row in rows:
    print "   ", row[1], "  \t", row[0]
```

```python
# analysys:
# -- retweets per user--

query = """select author_id, count(*) as xyz
FROM public.re_tweet
group by author_id
order by xyz desc"""

cur.execute(query)
rows = cur.fetchall()

print "Number of retweets:   \tAuthorId:"
for row in rows:
    print "   ", row[1], "  \t", row[0]
```

```python
# analysys:
# -- authors number per retweets number--

query = """select sub.num_of_retweets, count(*) as authors_number
from (
select author_id, count(*) as num_of_retweets
FROM public.re_tweet
group by author_id) as sub
group by sub.num_of_retweets
order by sub.num_of_retweets asc"""

cur.execute(query)
rows = cur.fetchall()

print "Number of authors:   \tNumber of retweets:"
for row in rows:
    print "   ", row[1], "  \t\t", row[0]
```

```python
# analysys:
# -- retweets  per date--

query = """select date, count(*) as retweets from
re_tweet group by date
order by date asc"""

cur.execute(query)
rows = cur.fetchall()

print "Date:   \tNumber of retweets:"
for row in rows:
    print "   ", row[0], "  \t\t", row[1]
```

```python
# analysys:
# -- tweets  per date--

query = """select substring(date, 1, 10) as kiedy, count(*) as tweets_per_date
from tweet
group by kiedy
order by kiedy"""

cur.execute(query)
rows = cur.fetchall()

print "Date:   \tNumber of tweets:"
for row in rows:
    print "   ", row[0], "  \t\t", row[1]
```

```python
# analysys:
# -- retweets per user--

query = """select author.name, count(*) as xyz
FROM public.re_tweet
left join author on author_id = author.id
group by author.name
order by xyz desc"""

cur.execute(query)
rows = cur.fetchall()

print "Number of retweets:   \tAuthorId:"
for row in rows:
    print "   ", row[1], "  \t", row[0]
```

```python
# analysys:
# -- retweets per user--

query = """select author.name, author_id, count(*) as xyz
FROM public.re_tweet
left join author on author_id = author.id
group by author.name, author_id
order by xyz desc"""

cur.execute(query)
rows = cur.fetchall()

print "Number of retweets:   \tAuthorName:   \t\tAuthorId"
for row in rows:
    print "   ", row[2], "  \t\t", row[0], "  \t\t\t", row[1]
```
